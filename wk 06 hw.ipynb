```python
The simple linear regression model is:

ùëå=ùõΩ0+ùõΩ1 ùëã+ùúñ 
where:

ùõΩ0 is the intercept, the value of ùëåwhen ùëã=0
ùõΩ1 is the slope, showing the change in Y for a one-unit increase in X.
œµ is the error term, representing unexplained random variation.
For a given X, Y is normally distributed around the predicted value, with the error term controlling its variability.
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)  # For reproducibility
X = np.linspace(0, 10, 50)  # Independent variable
true_slope = 2.5
true_intercept = 1.5
noise = np.random.normal(0, 1, X.shape)
Y = true_slope * X + true_intercept + noise  # Dependent variable

# Perform linear regression using scipy.stats
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

# Print the estimated parameters and goodness of fit
print("Estimated Slope (Œ≤1):", slope)
print("Estimated Intercept (Œ≤0):", intercept)
print("R-squared:", r_value**2)
print("P-value:", p_value)
print("Standard Error:", std_err)

# Define the regression line
Y_pred = slope * X + intercept

# Plot the data points and regression line
plt.scatter(X, Y, color="blue", label="Data Points")
plt.plot(X, Y_pred, color="red", label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set seed for reproducibility
np.random.seed(0)

# Simulate data based on a simple linear model
X = np.linspace(0, 10, 50)
true_slope = 2.5
true_intercept = 1.5
noise = np.random.normal(0, 1, X.shape)
Y = true_slope * X + true_intercept + noise

# Create a pandas DataFrame
data = pd.DataFrame({
    'X': X,
    'Y': Y
})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Print the model summary
print(model.summary())

# Visualize the data and the fitted regression line
plt.scatter(data['X'], data['Y'], label='Data Points', color='blue')
plt.plot(data['X'], model.fittedvalues, color='red', label='Fitted Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression with Statsmodels")
plt.legend()
plt.show()
```

```python
In a fitted Simple Linear Regression model, the in-sample predictions (or fitted values) are derived from the estimated parameters of the model. When you call fitted_model.fittedvalues in statsmodels, it provides these predicted 
ùëå
Y values using the estimated coefficients (or parameters) found in fitted_model.params or displayed in fitted_model.summary().tables[1].
```

```python
In the "ordinary least squares" (OLS) method, the line chosen for the fitted model is the one that minimizes the sum of the
squared differences (or "squares") between the observed Y values and the predicted Y values from the line. 
```

```python
The first formula calculates ùëÖ^2,the proportion of Y‚Äôs total variation explained by the model‚Äôs predictions (fitted_model.fittedvalues). 
A higher ùëÖ^2 means the model explains more variation in Y, indicating better accuracy.
```

```python

In the context of a Simple Linear Regression model, the null hypothesis of "no linear association (on average)" between 
the independent variableX (predictor) and the dependent variable Y (outcome) can be specified as:

ùêª0:ùõΩ1=0
where:
ùõΩ1 is the slope of the regression line, representing the average change in Y for a one-unit increase in X.
Under ùêª0, if ùõΩ1=0 Ôºåit implies that changes in X do not affect Y on average, indicating no linear association between X and Y.
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Filter long wait times from the Old Faithful dataset
long_wait_limit = 70  # Define the threshold for "long wait times"
long_wait_data = old_faithful[old_faithful['waiting'] >= long_wait_limit]
n = len(long_wait_data)

# Number of bootstrap samples
n_bootstraps = 1000
bootstrap_slopes = []

# Bootstrap sampling to estimate slope distribution
for _ in range(n_bootstraps):
    bootstrap_sample = long_wait_data.sample(n, replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Plot the bootstrapped sampling distribution of the slope coefficients
plt.hist(bootstrap_slopes, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(np.mean(bootstrap_slopes), color='red', linestyle='--', label="Mean Bootstrap Slope")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.title("Bootstrap Sampling Distribution of Slope Coefficients")
plt.legend()
plt.show()

# Calculate the 95% confidence interval from the bootstrap slopes
bootstrap_ci = np.percentile(bootstrap_slopes, [2.5, 97.5])
print(f"95% Bootstrap Confidence Interval for Slope: {bootstrap_ci}")

# Check if 0 is within the 95% confidence interval
contains_zero = bootstrap_ci[0] <= 0 <= bootstrap_ci[1]
print(f"Does the CI contain 0? {'Yes' if contains_zero else 'No'}")

# Simulate under the null hypothesis (no association, slope = 0)
null_slope = 0
null_intercept = long_wait_data['duration'].mean()  # Mean as intercept under null
null_slopes = []

for _ in range(n_bootstraps):
    # Generate simulated 'duration' data under null hypothesis
    simulated_duration = null_intercept + null_slope * long_wait_data['waiting'] + np.random.normal(0, 1, n)
    simulated_data = pd.DataFrame({'waiting': long_wait_data['waiting'], 'duration': simulated_duration})
    
    # Fit model on simulated data and collect slope
    null_model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    null_slopes.append(null_model.params['waiting'])

# Plot the null sampling distribution of the slope coefficients
plt.hist(null_slopes, bins=30, alpha=0.7, color='salmon', edgecolor='black')
plt.axvline(0, color='red', linestyle='--', label="Null Hypothesis Slope (0)")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.title("Null Sampling Distribution of Slope Coefficients")
plt.legend()
plt.show()

# Calculate simulated p-value: proportion of bootstrap slopes as extreme as the observed slope
observed_slope = smf.ols('duration ~ waiting', data=long_wait_data).fit().params['waiting']
p_value_simulated = np.mean([abs(slope) >= abs(observed_slope) for slope in null_slopes])
print(f"Simulated p-value: {p_value_simulated}")

# Compare with p-value from `statsmodels`
original_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
original_p_value = original_model.pvalues['waiting']
print(f"Original p-value from statsmodels: {original_p_value}")
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Original model to get the observed slope
observed_model = smf.ols('duration ~ waiting', data=old_faithful).fit()
observed_slope = observed_model.params['waiting']

# Permutation test
n_permutations = 1000
permuted_slopes = []

for _ in range(n_permutations):
    # Shuffle the duration values to break any association with waiting
    shuffled_duration = np.random.permutation(old_faithful['duration'])
    # Fit the model with shuffled durations
    permuted_model = smf.ols('shuffled_duration ~ waiting', 
                             data=old_faithful.assign(duration=shuffled_duration)).fit()
    # Collect the slope of waiting
    permuted_slopes.append(permuted_model.params['waiting'])

# Calculate the p-value as the proportion of permuted slopes as extreme as observed slope
p_value = np.mean([abs(slope) >= abs(observed_slope) for slope in permuted_slopes])

print(f"Observed Slope: {observed_slope}")
print(f"Permutation Test p-value: {p_value}")

# Plot the permuted slope distribution
import matplotlib.pyplot as plt
plt.hist(permuted_slopes, bins=30, color='skyblue', edgecolor='black')
plt.axvline(observed_slope, color='red', linestyle='--', label='Observed Slope')
plt.xlabel("Slope Coefficient (Permutation)")
plt.ylabel("Frequency")
plt.title("Permutation Test Distribution of Slope Coefficients")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Create an indicator variable for wait time length
old_faithful['wait_time_length'] = ['short' if x < 68 else 'long' for x in old_faithful['waiting']]

# Define groups
short_group = old_faithful[old_faithful['wait_time_length'] == 'short']['duration']
long_group = old_faithful[old_faithful['wait_time_length'] == 'long']['duration']

# Bootstrap resampling
n_bootstraps = 1000
bootstrapped_mean_differences = []

for _ in range(n_bootstraps):
    # Resample with replacement within each group
    short_sample = np.random.choice(short_group, size=len(short_group), replace=True)
    long_sample = np.random.choice(long_group, size=len(long_group), replace=True)
    
    # Calculate the mean difference between the two groups
    mean_difference = long_sample.mean() - short_sample.mean()
    bootstrapped_mean_differences.append(mean_difference)

# Calculate the 95% bootstrap confidence interval
ci_lower, ci_upper = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval for Mean Difference: [{ci_lower:.3f}, {ci_upper:.3f}]")

# Plot the distribution of bootstrap mean differences
import matplotlib.pyplot as plt
plt.hist(bootstrapped_mean_differences, bins=30, color='skyblue', edgecolor='black')
plt.axvline(ci_lower, color='red', linestyle='--', label='2.5th Percentile')
plt.axvline(ci_upper, color='red', linestyle='--', label='97.5th Percentile')
plt.xlabel("Mean Difference (Long - Short)")
plt.ylabel("Frequency")
plt.title("Bootstrap Distribution of Mean Differences")
plt.legend()
plt.show()
```

```python

For the bootstrap simulation, we repeatedly sample with replacement from each group (short and long wait times), calculating the mean difference in each bootstrap sample. This creates a distribution of mean differences, reflecting the variability expected due to random sampling within each group.

For the permutation test simulation, we randomly shuffle the duration values, breaking any association with waiting while preserving the original data structure. Each shuffle represents a sample under the null hypothesis, allowing us to build a distribution of slopes under the assumption of no relationship.
```
